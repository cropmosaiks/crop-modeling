```python
import numpy as np
import pandas as pd
```

```python
# satellite = "landsat-8-c2-l2"
# satellite = "sentinel-2-l2a"
satellite = "landsat-c2-l2"

country_code = 'ZMB' # 'TZA' # 'NGA'

# points = 15
points = 20

num_features = 1024

# bands_short = "2-3-4"
# bands_short = "2-3-4-8"
# bands_short = "1-2-3-4-5-6-7"
bands_short = "r-g-b-nir-swir16-swir22"
# bands_short = "2-3-4-5-6-7-8-11-12"

year = 2014

file_name = (f'{satellite}_bands-{bands_short}_{country_code}_{points}'+
             f'k-points_{num_features}-features_{year}')
file_name
```

```python
df_1 = pd.read_feather(f"partial_features/{file_name}_1.feather")
df_2 = pd.read_feather(f"partial_features/{file_name}_2.feather")
df_3 = pd.read_feather(f"partial_features/{file_name}_3.feather")
df_4 = pd.read_feather(f"partial_features/{file_name}_4.feather")
df_5 = pd.read_feather(f"partial_features/{file_name}_5.feather")
df_6 = pd.read_feather(f"partial_features/{file_name}_6.feather")
df_7 = pd.read_feather(f"partial_features/{file_name}_7.feather")
df_8 = pd.read_feather(f"partial_features/{file_name}_8.feather")
df_9 = pd.read_feather(f"partial_features/{file_name}_9.feather")
df_10 = pd.read_feather(f"partial_features/{file_name}_10.feather")
df_11 = pd.read_feather(f"partial_features/{file_name}_11.feather")
df_12 = pd.read_feather(f"partial_features/{file_name}_12.feather")
df_final = pd.concat([
    df_1, 
    df_2, 
    df_3,
    df_4,
    df_5,
    df_6,
    df_7,
    df_8,
    df_9,
    df_10,
    df_11,
    df_12
]).reset_index(drop = True)
df_final
```
