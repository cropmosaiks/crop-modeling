# Concatenate feature files

```python
import numpy as np
import pandas as pd
from pyhere import here
import glob
import os
```

```python
def get_merged_files(flist, **kwargs):
    return pd.concat([pd.read_feather(f, **kwargs) for f in flist], axis=0)
```

## Monthly files to yearly

```python
# satellite = "landsat-8-c2-l2"
# satellite = "sentinel-2-l2a"
satellite = "landsat-c2-l2"

country_code = 'ZMB' # 'TZA' # 'NGA'

# points = 15
points = 20

num_features = 1024

# bands_short = "2-3-4"
# bands_short = "2-3-4-8"
# bands_short = "1-2-3-4-5-6-7"
bands_short = "r-g-b-nir-swir16-swir22"
# bands_short = "2-3-4-5-6-7-8-11-12"

year = 2014

file_name = (f'{satellite}_bands-{bands_short}_{country_code}_{points}'+
             f'k-points_{num_features}-features_{year}')
file_name
```

```python
path = str(here('data', 'random_features', 'partial_features', f'{file_name}_*.feather'))
files = glob.glob(pathname=path)
df = get_merged_files(files)
df = df.sort_values(by = ['month']).reset_index(drop = True)
df
```
